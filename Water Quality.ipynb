```python
#Import pandas, matplotlib.pyplot, and seaborn in the correct lines below
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# the supplied CSV data file is the raw_data directory
water_data = pd.read_csv('C:/Users/monal/Desktop/Data Science/Water Quality/water_potability.csv')
```

```python
#1 is potable, 0 is nonpotable
water_data.dropna(inplace = True)
water_data.head()
```

```python
water_data.plot(x='Organic_carbon', y='Turbidity', kind='scatter');
```

```python
water_data.isnull().sum().sum()
```

```python
potable = water_data[water_data['Potability'] == 1]
nonpotable = water_data[water_data['Potability'] == 0]
```

```python
X = water_data.drop(columns = ['Potability'])
y = water_data['Potability']
print(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X , y, test_size = 0.2)
```

```python
model = DecisionTreeClassifier()
#model.fit(X, y)
#model.predict([[7, 210, 14000, 5, 310, 470, 15, 80, 4.5]])
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```python
print("Model accuracy:", accuracy_score(y_test, predictions))
```
